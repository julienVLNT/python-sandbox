```python
import math
from matplotlib import pyplot
import numpy
```

# Opérateurs de différences finies d'ordre élevé en 1.D

```python
def differences_centrees(nop : int, n: int) -> numpy.ndarray :
    """Calcul des coefficients de l'opérateur par différences finies centrées.

    PARAMETRES :
    ------------
    nop : int,
          entier naturel impair, nombre de points du stencil
    n   : int,
          entier naturel, ordre de dérivation
    
    RETOUR :
    --------
    operateur : numpy.ndarray
                coefficients de l'opérateur de différenciation
    """

    m = numpy.zeros((nop, nop))
    for i in range(nop):
        for j in range(nop):
            dx     = j - nop // 2
            m[i,j] = dx ** i
    
    s    = numpy.zeros(nop)
    s[n] = math.factorial(n)

    operateur = numpy.linalg.solve(m, s)
    return operateur
```

```python
def afficher(nop : int, operateur : numpy.ndarray) -> None:
    """ Affiche le stencil de l'opérateur. """
    
    x = numpy.linspace(-(nop - 1) / 2, (nop - 1) / 2, nop)

    pyplot.figure(figsize=(10, 4))
    pyplot.plot(x, operateur,lw=2,color='blue')
    pyplot.plot(x, operateur,lw=2,marker='o',color='blue')
    pyplot.plot(0, 0,lw=2,marker='o',color='red')

    pyplot.title("Opérateur de Taylor avec nop =  %i " % nop )
    pyplot.xlabel('x')
    pyplot.ylabel('Operator')
    pyplot.grid()
    pyplot.show()
```

## Interpolation : `n=0`

```python
nop = 3
operateur = differences_centrees(nop, 0)
afficher(nop, operateur)
```

## Dérivée première : `n=1`

```python
nop = 3
operateur = differences_centrees(nop, 1)
afficher(nop, operateur)
```

```python
nop = 5
operateur = differences_centrees(nop, 1)
afficher(nop, operateur)
```

```python
nop = 7
operateur = differences_centrees(nop, 1)
afficher(nop, operateur)
```
